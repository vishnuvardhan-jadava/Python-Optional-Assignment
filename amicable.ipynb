#amicable numbers in the given range

```python
def sum_of_divisors(n):
  """this function takes an integer as input and returns the sum of all the divisors"""
  divisors=[]
  for i in range(1,int(n)):
    if (n%i)==0:
      divisors.append(i)
  return sum(divisors)
#print(sum_of_divisors(38))
```

```python
def amicable(start,end):
  """this function takes returns the amiable numbers in the given range"""
  li=[]
  for i in range(int(start),int(end)):
    for j in range(int(start),int(end)):
      if i ==j:
        pass
      else:
        if ((sum_of_divisors(i)==j) and (sum_of_divisors(j)==i)) and (i not in li) and (j not in li):
          li.append(i),li.append(j)
          print(i, 'and',j,'are amicable')
amicable(2,1000)
```
