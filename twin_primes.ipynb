```python
def is_prime(n):
    count=0
    for i in range(2,int(n/2)+2):
        if int(n % i) == 0:
            count+=1
    if count>0:
        #print('not_prime')
        pass
    else:
        return n
is_prime(28)
```

```python
li=[x for x in range(2,1001)]
li = list(filter(lambda x: int(x%2)==1,li))
cc=[]
for i in li:
    for j in li:
        if (is_prime(i) and is_prime(j) and (j-i ==2)):
            print((i,j),'are twin-primes')
            cc.append(i);cc.append(j)
```
