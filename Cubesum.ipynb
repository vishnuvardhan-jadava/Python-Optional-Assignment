#cubesum function: to find the sum of cubes of given number

```python
def cubesum(n):
  """This function returns the sum of cubes of individual digits of a given number"""
  s=0
  while n>1:
    s += (int(n%10)**3)
    n/=10
  return s
print(cubesum(int(153.4)))
```

#isArmstrong: to find whether the given number is an Armstrong number or not

```python
def isArmstrong(num):
  """this function takes input as an integer and prints whether that number is armstrong or not."""
  print('Armstrong') if num == cubesum(num) else print('Not Armstrong')
isArmstrong(int(153))
```

#PrintArmstrong: to print Armstrong numbers till the given number

```python
def PrintArmstrong(num):

  """
  This function takes an interger as input and prints the armstrong and non-armstrong number till the given number 
  """
  for i in range(1,num+1):
    print(i,' is Armstrong') if i == cubesum(i) else print(i,' is Not Armstrong')
PrintArmstrong(int(200))
```
