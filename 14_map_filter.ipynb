---> To check whether a number is even or not, you need to use filter() function.

---> To generate a cube of an even number, you need to use map() function.

```python
x=list(filter(lambda x:int(x%2==0),[x for x in range(1,29)]))
x
```

```python
y_map = list(map(lambda x:x**3,list(filter(lambda x:int(x%2==0),[x for x in range(1,29)]))))
y_map
```
