```python
def sumPdivisors(n):
  """this function takes an interger and returns sum of all the divisors"""
  li=[]
  for i in range(1,n):
    if (n % i) ==0:
      li.append(i)
  #print('divisiors of {} are: {}'.format(n,li))
  return sum(li)

print(sumPdivisors(int(26)))
```
