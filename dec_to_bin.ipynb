```python
def dec_to_bin(n):
  """this function takes decimal number as input and converts it into a binary number"""
  li=[]
  while n:
    r = str(n%2)
    li.append(r)
    n=int(n/2)
  #print(li[::-1])
  return int(''.join(li[::-1]))
dec_to_bin(9)
```
