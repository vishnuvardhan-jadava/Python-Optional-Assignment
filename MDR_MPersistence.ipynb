```python
def prodDigits(n):
  """this function takes input as an interger and returns the product of individual digits"""
  s=1
  while n >1:
    s *= int(n%10)
    n/=10
  return s
#print(prodDigits(int(12345)))
```

```python
def mdr(n):
  """this function takes input as an interger and returns multiplicative digital root"""
  val = int(n)
  lenn =len(str(val))

  while lenn > 1:
    val=int(prodDigits(val))
    lenn=len(str(val))
  #print('val is {} and len is {}'.format(val,lenn))
  print(f'MDR of {n} is: {val}')
mdr(84)
```

```python
def MPersistence(n):
  """this function takes input as an interger and returns Multipliative Persistene"""
  val = int(n)
  lenn =len(str(val))
  count=0
  while lenn > 1:
    val=int(prodDigits(val))
    lenn=len(str(val))
    #print('val is {} and len is {}'.format(val,lenn))
    count+=1
  print('Multipliative Persistene of {} is: {} '.format(n,count))
MPersistence(86)
```
