```python
def fact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)
#fact(6)
```

```python
def permutations(n,r):
  """this function takes n and r and returns permutations (p(n, r) = n! / (n-r)!)"""
  return (fact(n)/fact(n-r))
permutations(10,3)
```

```python
def combinations(n,r):
  """this function takes n and r and returns combinations(c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!)"""
  print((fact(n)/(fact(r)*fact(n-r))))
  #print(permutations(n,r)/fact(r))
combinations(10,3)
```
